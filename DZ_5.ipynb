```python
#ИСХОДНЫЕ ДАННЫЕ

#Перечень всех документов
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

#Перечень полок, на которых хранятся документы
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
#ЗАДАНИЕ 1

#Функция поиска владельца по номеру документа
#Входные параметры: перечень всех документов, номер для поиска  
#Выводит имя владельца или сообщение, если документ не найден
def name_from_number(documents, doc_number):
    for doc in documents:
        if doc['number'] == doc_number:
            print('Владелец документа: ' + doc['name'])
            return
    print('Документ не найден в базе')
    return

#Функция поиска места хранения по номеру документа
#Входные параметры: перечень полок с документами, номер для поиска  
#Выводит место хранения или сообщение, если документ не найден
def shelf_from_number(directories, doc_number):
    for shelf in directories:
        if doc_number in directories[shelf]:
            print('Документ хранится на полке: ' + shelf)
            return
    print('Документ не найден в базе')
    return

#Функция вывода на экран полной информации по всем документам
#Входные параметры: перечень всех документов, перечень полок с документами 
#Выводит общий список всех документов
def full_information(documents, directories):
    for doc in documents:
        print('№: ' + doc['number'] + ', тип: ' + doc['type'] + ', владелец: ' + doc['name'], end = '')
        for shelf in directories:
            if doc['number'] in directories[shelf]:
                print(', полка хранения: ' + shelf)
    return
        
#Функция добавления новой полки
#Входной параметр: перечень полок с документами, номер полки
#Выводит сообщение об успешном добавлении полки или что полка уже существует
def shelf_add(directories, shelf_number):
    if shelf_number in directories:
        print('Такая полка уже существует. ' + shelf_list(directories))
    else:
        directories[shelf_number] = []
        print('Полка добавлена. ' + shelf_list(directories))
    return

#Функция вывода перечня полок
#Входной параметр: перечень полок с документами
#Возвращает перечень полок по возрастанию номера
def shelf_list(directories):
    shelf_list_int = sorted(int(i) for i in directories.keys())
    return('Текущий перечень полок: ' + ', '.join(str(i) for i in shelf_list_int) + '.')

#Функция удаления пустой полки
#Входной параметр: перечень полок с документами, номер полки
#Выводит сообщение об успешном удалении пустой полки или что на полке есть документ или что полки не существует
def shelf_del(directories, shelf_number):
    if shelf_number in directories and directories[shelf_number] == []:
        directories.pop(shelf_number)
        print('Полка удалена. ' + shelf_list(directories))
    elif shelf_number in directories and directories[shelf_number] != []:
        print('На полке есть документы, удалите их перед удалением полки. ' + shelf_list(directories))
    else:
        print('Такой полки не существует. ' + shelf_list(directories))
    return
```

```python
#ЗАДАНИЕ 2

#Функция добавления нового документа
#Входные параметры: перечень всех документов, перечень полок с документами 
#Возвращает сообщение об успешном добавлении документа или что полки не существует
def document_add(documents, directories):
    doc_number = input('Введите номер документа:\n')
    doc_type = input('Введите тип документа:\n')
    doc_name = input('Введите владельца документа:\n')
    doc_shelf = input('Введите полку для хранения:\n')
    if  doc_shelf in directories:
        documents.append({'type': doc_type, 'number': doc_number, 'name': doc_name})
        directories[doc_shelf] = doc_number
        print('Документ добавлен. Текущий список документов:')
        full_information(documents, directories)
    else:
        print('Такой полки не существует. Добавьте полку командой as.\nТекущий список документов:')
        full_information(documents, directories)
    return    

#Функция удаления документа по номеру
#Входные параметры: перечень всех документов, перечень полок с документами, номер для поиска
#Возвращает сообщение об успешном удалении документа или что документа не существует
def document_del(documents, directories, doc_number):
    for doc in documents:
        if doc['number'] == doc_number:
            documents.remove(doc)
            for shelf in directories:
                if doc_number in directories[shelf]:
                    directories[shelf].remove(doc_number)
            print('Документ удален.\nТекущий список документов:')
            full_information(documents, directories)
            return
    print('Документ не найден в базе.\nТекущий список документов:')
    full_information(documents, directories)
    return

#Функция перемещения документа с полки на полку
#Входные параметры: перечень всех документов, перечень полок с документами, номер документа, номер полки
#Возвращает сообщение об успешном перемещении документа или что документа не существует или что полки не существует
def document_remove(documents, directories, doc_number, shelf_number):
    for doc in documents:
        if doc['number'] == doc_number and shelf_number in directories:
            for shelf in directories:
                if doc_number in directories[shelf]:
                    directories[shelf].remove(doc_number)
            directories[shelf_number].append(doc_number)
            print('Документ перемещен.\nТекущий список документов:')
            full_information(documents, directories)
            return
    if doc_number not in [doc['number'] for doc in documents]:
        print('Документ не найден в базе.\nТекущий список документов:')
        full_information(documents, directories)
    if shelf_number not in directories:
        print('Такой полки не существует. ' + shelf_list(directories))
    return
```

```python
#ОСНОВНАЯ ПРОГРАММА

#Функция по команде пользователя вызывает соответствующую функцию
#Входной параметр: команда пользователя
#Возвращает принятую команду
def select_function(user_input):
    if user_input == 'p':
        name_from_number(documents, input('Введите номер документа:\n'))
    elif user_input == 's':
        shelf_from_number(directories, input('Введите номер документа:\n'))
    elif user_input == 'l':
        full_information(documents, directories)
    elif user_input == 'ads':
        shelf_add(directories, input('Введите номер полки:\n'))
    elif user_input == 'ds':
        shelf_del(directories, input('Введите номер полки:\n'))
    elif user_input == 'ad':
        document_add(documents, directories)
    elif user_input == 'd':
        document_del(documents, directories, input('Введите номер документа:\n')) 
    elif user_input == 'm':
        document_remove(documents, directories, input('Введите номер документа:\n'), input('Введите номер полки:\n'))
    return(user_input)

#Пользовательский ввод обрабатывается в цикле while до тех пор, 
#пока пользователь явно не завершит программу (вводом команды "q")
while select_function(input('Введите команду:\n')) != 'q':
    print('\n')
```
